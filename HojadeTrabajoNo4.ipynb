### Sección 2: ¿Cuál es el mes con mejor venta?

#### Se procede a eliminar las filas en blanco

```python
datacomplete2 = datacomplete.dropna()
```

```python
datacomplete2.head()
```

#### Se obtiene el numero de mes a partir de la fecha

```python
datacomplete2 = datacomplete2[datacomplete2['Order Date'].str[0:2] !='Or']
```

```python
datacomplete2['Month'] = datacomplete2['Order Date'].str[0:2]
```

```python
datacomplete2['Month'] = pd.to_numeric(datacomplete2['Month'], downcast="integer")
```

```python
datacomplete2.head()
```

#### Se procede a calcular los ingresos

```python
datacomplete2['Quantity Ordered'] = pd.to_numeric(datacomplete2['Quantity Ordered'], downcast="integer")
```

```python
datacomplete2['Price Each'] = pd.to_numeric(datacomplete2['Price Each'], downcast="float")
```

```python
datacomplete2['Sales'] = datacomplete2['Quantity Ordered']*datacomplete2['Price Each']
```

```python
datacomplete2.head()
```

#### Se filtra la información para que únicamente quede 2019

```python
import datetime

datacomplete2['year'] = pd.DatetimeIndex(datacomplete2['Order Date']).year
```

```python
datacomplete2.head()
```

```python
datacomplete2 = datacomplete2[datacomplete2['year'] == 2019] 
```

#### Se agrupa la información de ingresos por mes

```python
datacomplete2.groupby(['Month'])[['Sales']].sum()
```

De la tabla anterior se puede observar que el mes con más ingresos es diciembre.

### Sección 3: ¿Cuál es la cuidad que tiene mayor número de ventas?

#### Separación de la dirección para obtener la cuidad y el estado

```python
datacomplete2['Purchase Address'] = datacomplete2['Purchase Address'].str.split(',')
```

```python
datacomplete2.head()
```

```python
datacomplete2['City'] = datacomplete2['Purchase Address'].apply(lambda x: x[1].strip())
```

```python
datacomplete2.head()
```

```python
datacomplete2['State'] = datacomplete2['Purchase Address'].apply(lambda x: x[2].strip())
```

```python
datacomplete2.head()
```

```python
datacomplete2['State'] = datacomplete2['State'].str[0:2]
```

```python
datacomplete2.head()
```

```python
datacomplete2['CityState'] = datacomplete2['City']+', '+datacomplete2['State']
```

```python
datacomplete2.head()
```

```python
datacomplete2.groupby(['CityState'])[['Sales']].sum()
```

La cuidad con mayores ingresos es San Francisco, California.
